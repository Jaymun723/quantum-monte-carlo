```python
from production import Problem, ExactSolver
import numpy as np
```

```python
pb = Problem(n_sites=8, J_x=2.2, J_z=1.0, temperature=1.0, m=2)
```

```python
def S_i(pb: Problem, i: int):
    n = pb.n_sites
    ops = [np.eye(2, dtype="complex128") for _ in range(n)]
    ops[i] = 0.5 * np.array([[1, 0], [0, -1]], dtype="complex128")
    S = ops[0]
    for op in ops[1:]:
        S = np.kron(S, op)
    return S
```

```python
W_z = np.zeros((1 << pb.n_sites, 1 << pb.n_sites))
for i in range(pb.n_sites):
    W_z + S_i(pb, i)
```

```python
solver = ExactSolver(pb)
solver.compute(solver.H, n_up=pb.n_sites // 2)
```
